```python
import pandas as pd
import zipfile
import os
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hmavrodiev/london-bike-sharing-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(os.path.join(path, "london_merged.csv"))
df.head()
```

```python
new_columns_dict = {
    "timestamp":"time",
    "cnt":"count",
    "t1":"temp_real_C",
    "t2":"temp_feels_like_C",
    "hum":"humidity_percent",
    "wind_speed":"wind_speed_kph",
    "weather_code":"weather",
    "is_holiday":"is_holiday",
    "is_weekend":"is_weekend",
    "season":"season"
}

df.rename(new_columns_dict,axis=1,inplace=True)
```

```python
df
```

```python
#make humidity value to be between 0 and 1
df.humidity_percent = df.humidity_percent/100
```

```python
#name the seasons
seasons_dict={
    "0.0":"spring",
    "1.0":"summer",
    "2.0":"autumn",
    "3.0":"winter"
}

#converting the values in season from float to string
df.season=df.season.astype("str")
#mapping these values/ replacing
df.season=df.season.map(seasons_dict)
```

```python
df.season
```

```python
weather_dict={
    "1.0":"Clear",
    "2.0":"Scattered clouds",
    "3.0":"Broken clouds",
    "4.0":"clouds",
    "7.0":"Rain",
    "10.0":"Rain with thunderstorm",
    "26.0":"Snowfall"
}

df.weather=df.weather.astype("str")
df.weather=df.weather.map(weather_dict)
```

```python
df.weather
```

```python
df.to_excel('london_bikes_final.xlsx',sheet_name='Data')
```
